```python
import geopandas as gpd
from lonboard import viz, Map, ScatterplotLayer
import numpy as np
import fsspec
```

```python
url = 'https://data.source.coop/boston-university/bu-glance/bu_glance_training_dataV2.parquet'

with fsspec.open(url) as file:
    gdf = gpd.read_parquet(file)
```

```python
viz(gdf)
```

```python
gdf.columns
```

```python
gdf.LC_Class.unique()
```

```python
color_dict = {'Trees':[53,130,33,255],#'#358221', 
              'Herbaceous':[171, 200, 53,255],#'#ABC835', 
              'Water':[30, 157, 179,255],#'#1e9db3', 
              'Shrub':[178, 159, 61,255],#'#B29E3D', 
              'Developed':[195, 109, 169,255],#'#C36DA9',
              'Barren/sparsely vegetated':[165, 155, 143 ,255],#'#EDE9E4', 
              'Ice/snow':[203, 241, 241,255]}#'#CBF1F1'}
```

```python
layer = ScatterplotLayer.from_geopandas(gdf)
map_ = Map(layers=[layer], _height=900)
map_
```

```python
layer.opacity = 0.9
layer.radius_scale = 100
layer.radius_min_pixels=0.5
layer.get_radius = 2
layer.get_fill_color = np.array(gdf.LC_Class.map(color_dict).values.tolist()).astype('uint8')
```
